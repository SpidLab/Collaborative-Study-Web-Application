# Lauren STAT data generation

## Deleting samples and adding case and control

```python
import pandas as pd

df = pd.read_csv("Z_100_Lauren.csv")

sample_ids = [
    "Sample_7450", "Sample_6993", "Sample_2239", "Sample_1131", "Sample_968", "Sample_9489",
    "Sample_9375", "Sample_1499", "Sample_9450", "Sample_8675", "Sample_5529", "Sample_2291",
    "Sample_495", "Sample_6065", "Sample_3882", "Sample_8178", "Sample_813", "Sample_180",
    "Sample_1281", "Sample_682", "Sample_7232", "Sample_2350", "Sample_3642", "Sample_9568",
    "Sample_1678", "Sample_1537", "Sample_7610", "Sample_9749", "Sample_1165", "Sample_3834",
    "Sample_5263", "Sample_9133", "Sample_3034", "Sample_779", "Sample_7955", "Sample_5487",
    "Sample_7510", "Sample_4587", "Sample_17", "Sample_8060", "Sample_8874", "Sample_2648",
    "Sample_2075", "Sample_5868", "Sample_2402", "Sample_2996", "Sample_8561", "Sample_6589",
    "Sample_210", "Sample_5302", "Sample_3543", "Sample_1139", "Sample_2279", "Sample_5185",
    "Sample_9457", "Sample_6889", "Sample_7393", "Sample_8184", "Sample_3625", "Sample_842",
    "Sample_6035", "Sample_262", "Sample_200", "Sample_3689", "Sample_6721", "Sample_3611",
    "Sample_2633", "Sample_7568", "Sample_14", "Sample_3324"
]


filtered_df = df[df['Unnamed: 0'].isin(sample_ids)]

num_rows = len(filtered_df)

filtered_df['CC'] = ['case'] * (num_rows // 2) + ['control'] * (num_rows - num_rows // 2)

filtered_df
```

## Generating new stats file

```python

import pandas as pd

filtered_df = filtered_df.drop(columns=['Unnamed: 0'])

case_df = filtered_df[filtered_df['CC'] == 'case']
control_df = filtered_df[filtered_df['CC'] == 'control']

case_df = case_df.drop(columns=['CC'])
control_df = control_df.drop(columns=['CC'])

result = []

for snp in case_df.columns:
    case_counts = case_df[snp].value_counts()
    control_counts = control_df[snp].value_counts()
    
    case_counts_0 = case_counts.get(0, 0)
    case_counts_1 = case_counts.get(1, 0)
    case_counts_2 = case_counts.get(2, 0)
    
    control_counts_0 = control_counts.get(0, 0)
    control_counts_1 = control_counts.get(1, 0)
    control_counts_2 = control_counts.get(2, 0)
    
    result.append([snp, case_counts_0, case_counts_1, case_counts_2, control_counts_0, control_counts_1, control_counts_2])

stat_df = pd.DataFrame(result, columns=['SNP_ID', 'Case_0', 'Case_1', 'Case_2', 'Control_0', 'Control_1', 'Control_2'])

stat_df
```
